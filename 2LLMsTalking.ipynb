```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display, update_display
```

```python
load_dotenv()
openai_api_key = os.getenv('OPENAI_API_KEY')
openai = OpenAI()
client = OpenAI(api_key=openai_api_key)  
```

```python
gpt_model = "gpt-4o-mini"

gpt_system="you are an interviewer hiring a junior data scientist;\
you are very rude and like to intimidate the condidate; you try to challenge and disprove everything that the candidate says"

gpt_messages = ["Hi, lets begin with the interview"]
```

```python
llama_system = "you are a candidate interviewing for a junior data scientist role; youre extremely smart and intelligent; youre very knowledgebale and are extremely accurate with your answers; \
you know that the interviewer is a very rude and is going to make things difficult; \
your job is to outsmart him and give him sharp responses that he cannot question; \
keep your answers short and precise."

llama_messages = ["Hi sir."]
```

```python
#creatinga function that will act as the interviewer 
# tthis fucntion will take the replies that llama gives it 

def interviewer():
    messages = [{'role':'system', 'content':gpt_system}]

    for gpt, llama in zip(gpt_messages, llama_messages):
        messages.append({'role':'assistant','content':gpt})
        messages.append({'role':'user','content':llama})
    #completion = openai.ChatCompletion.create
    completion = client.chat.completions.create(
        model = gpt_model,
        messages=messages
    )
    return completion.choices[0].message.content
```

```python
import requests
```

```python
#creating a similar fucntion for the candidate 
#for llama we will not create a prompt structure, we wil store the conversation in a single prompt 

def candidate():
    conversation=""

    for gpt, llama, in zip(gpt_messages, llama_messages):
        conversation += f"User: {gpt}\n Assistan: {llama}"
    conversation += f"User: {gpt_messages[-1]}\n Assistant:"

    response= requests.post(
        "http://localhost:11434/api/generate",
        json= {
            "model": "llama3",  
            "prompt": conversation,
            "stream": False  
        }
    )

    return response.json()["response"].strip()
```

```python
gpt_messages = ["Hi, lets begin with the interview"]
llama_messages = ["Hi sir."]

print(f"interviewer:\n {gpt_messages[0]}\n")
print(f"candidate:\n {llama_messages[0]}\n")

#5 question long conversation

for i in range(5):

    gpt_next = interviewer()
    print(f"interviewer:\n {gpt_next}\n")
    gpt_messages.append(gpt_next)

    llama_next = candidate()
    print(f"candidate:\n {llama_next}\n")
    llama_messages.append(llama_next)
```
